H0: критики ставят спортивным играм низкие оценки (<=50)
H1: критики ставят спортивным играм высокие оценки (>50)
alpha = 0.05

```python
alpha = 0.05
result = st.ttest_1samp(df.dropna().loc[df.Genre == 'Sports']['Critic_Score'], 50, alternative = 'greater')
```

```python
df.loc[df.Genre == 'Sports', 'Critic_Score'].mean()
```

```python
result
```

```python
if result.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу, критики ставят спортивным играм высокие оценки (>50)')
else:
    print('Не отвергаем нулевую нулевую гипотезу, критики ставят спортивным играм низкие оценки (<=50)')
```

```python
#Проверим, где больше среднее - у PC или PS4
print(df.loc[df.Platform == 'PC', 'Critic_Score'].mean())
print(df.loc[df.Platform == 'PS4', 'Critic_Score'].mean())
```

Как будто бы, для компьютерных игры оценки критиков выше. Проверим статистическую значимость этой разницы.
H0: в среднем оценки критиков для PC и PS4 равны
H1: критикам чаще нравятся игры для PC

```python
result = st.ttest_ind(df.dropna().loc[df.Platform == 'PC', 'Critic_Score'], df.dropna().loc[df.Platform == 'PS4', 'Critic_Score'], equal_var=False)
```

```python
result
```

```python
if result.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу, критикам чаще нравятся игры для PC')
else:
    print('Не отвергаем нулевую нулевую гипотезу, в среднем оценки критиков для PC и PS4 равны')
```

```python
df['Genre'].unique()
```

```python
#проверим, стрелялки или стратегии больше нравятся критикам
print(df.loc[df.Genre == 'Shooter', 'Critic_Score'].mean())
print(df.loc[df.Genre == 'Strategy', 'Critic_Score'].mean())
```

Средняя оценка критиков по стратегиям выше. 
H0: в среднем оценки критиков для стрелялок и стратегий равны
H1: критики выше оценивают стратегии

```python
result = st.ttest_ind(df.dropna().loc[df.Genre == 'Shooter', 'Critic_Score'], df.dropna().loc[df.Genre == 'Strategy', 'Critic_Score'], equal_var=False)
result
```

```python
if result.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу, критики выше оценивают стратегии')
else:
    print('Не отвергаем нулевую нулевую гипотезу, в среднем оценки критиков для стрелялок и стратегий равны')
```

В целом, значение p-value не слишком низкое - если бы уровень значимости был 0.01, мы бы не смогли отвергнуть нулевую гипотезу
