```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 11,7
```

```python
df = pd.read_csv('C:\\Users\\stolb\\Desktop\\Материалы к лекциям\\Статистика в Python\\Лекция 2\\archive\\penguins_lter.csv')
df
```

```python
df = df[['Sample Number', 'Species', 'Island', 'Date Egg', 'Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)', 'Sex']].dropna()
df
```

```python
df['Date Egg'] = pd.to_datetime(df['Date Egg'], format = '%m/%d/%y')
```

```python
df
```

```python
species = df['Species'].value_counts()
species.plot(kind='pie', title='Виды пингвинов')
```

Вывод: пингвины Адели встречаются в изученной области чаще остальных

```python
females07 = df.loc[(df.Sex == 'FEMALE') & (df['Date Egg'].dt.year == 2007)].groupby('Date Egg').count()[['Sample Number']].reset_index()
females07['Date Egg'] = females07['Date Egg'].dt.strftime('%m-%d')
females07['Date'] = pd.to_datetime(females07['Date Egg'], format = '%m-%d')
females07
```

```python
females08 = df.loc[(df.Sex == 'FEMALE') & (df['Date Egg'].dt.year == 2008)].groupby('Date Egg').count()[['Sample Number']].reset_index()
females08['Date Egg'] = females08['Date Egg'].dt.strftime('%m-%d')
females08['Date'] = pd.to_datetime(females08['Date Egg'], format = '%m-%d')
females07
```

```python
females09 = df.loc[(df.Sex == 'FEMALE') & (df['Date Egg'].dt.year == 2009)].groupby('Date Egg').count()[['Sample Number']].reset_index()
females09['Date Egg'] = females09['Date Egg'].dt.strftime('%m-%d')
females09['Date'] = pd.to_datetime(females09['Date Egg'], format = '%m-%d')
females09
```

```python
plt.plot(females07['Date'], females07['Sample Number'], label = '2007 год')
plt.plot(females08['Date'], females08['Sample Number'], label = '2008 год')
plt.plot(females09['Date'], females09['Sample Number'], label = '2009 год')
plt.title('Динамика откладывания яиц пингвинами')
plt.xlabel('Период')
plt.ylabel('яйца,шт')
plt.legend()
```

На графиках можно сравнить, как пингвины откладывали яйца в зависимости от года (пришлось привести их к одному году, иначе непонятно, как их разместить на оси поверх друг друга). Не самая удачная визуализация, но это единственное, что можно привязать ко времени в этом датасете. 
Вывод:
Мы видим, что в 2008 году период откладывания яиц смещен на более ранний срок, что, возможно, может говорить о более раннем наступлении лета в этом году. 
Также в 2007 году было всего два "пиковых" дня по откладыванию яиц, в остальные 2 года процесс шел более равномерно. О причинах остается только догадываться)))

```python
#можно было сделать через цикл, но не разобралась, каким образом тогда настроить разные цвета. По умолчанию он один.
adelie = df.loc[df.Species == 'Adelie Penguin (Pygoscelis adeliae)']
chinstrap = df.loc[df.Species == 'Chinstrap penguin (Pygoscelis antarctica)']
gentoo = df.loc[df.Species == 'Gentoo penguin (Pygoscelis papua)']
```

```python
sns.histplot(adelie['Body Mass (g)'], alpha = 0.5, color = 'red', label = 'Пингвин Адели', stat = 'density')
sns.histplot(chinstrap['Body Mass (g)'], alpha = 0.5, color = 'green', label = 'Антарктический пингвин', stat = 'density')
sns.histplot(gentoo['Body Mass (g)'], alpha = 0.5, color = 'blue', label = 'Папуанский пингвин', stat = 'density')
plt.title('Размеры пингвинов по видам')
plt.legend()
```

Вывод: Папуанские пингвины - более крупные по сравнению с остальными видами пингвинов, а размеры наблюдаемых антарктических пингвинов обладают наименьшим разбросом. 

```python
plt.scatter(x=adelie['Culmen Length (mm)'], y = adelie['Flipper Length (mm)'], label = 'Пингвин Адели' , color = 'red' ) 
plt.scatter(x=chinstrap['Culmen Length (mm)'], y = chinstrap['Flipper Length (mm)'], label = 'Антарктический пингвин', color = 'green', marker='D')
plt.scatter(x=gentoo['Culmen Length (mm)'], y = gentoo['Flipper Length (mm)'], label = 'Папуанский пингвин', color = 'blue', marker='X')
plt.title('Взаимосвязь между размерами клюва и длиной крыла пингвина')
plt.legend()
plt.xlabel('Длина клюва, мм')
plt.ylabel('Длина крыла, мм')
```

Выводы: в целом, наблюдается взаимосвязь по всем видам пингвинов - чем больше длина клюва, тем больше будет длина крыла. Однако, есть особь с длинным клювом и маленькими крыльями среди Антарктических пингвинов. Среди пингвинов Адели эта взаимосвязь наименее выражена. А также мы можем заметить, что у Антарктических пингвинов в среднем более длинные клювы относительно общих пропорций тела, чем у остальных видов.
