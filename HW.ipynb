Выбросы в столбцах respiratory_rate, rectal_temperature, pulse кажутся естественными, поэтому заменять или удалять их не будем, просто будем считать меры центральной тенденции без них.

```python
def remove_outliers(column,df_named):
    '''
    удаляет выбросы в указанном столбце
    '''
    q1 = df_named[column].quantile(0.25)
    q3 = df_named[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    df_named = df_named[df_named[column].between(lower_bound, upper_bound, inclusive = True)]
    return df_named
```

```python
df_without_outliers = remove_outliers('respiratory_rate', df_named)
df_without_outliers.describe()
```

```python
# respiratory_rate вообще считается сомнительным показателем, поэтому без зазрений совести и группировок заполним пропуски средним без учета выбросов
df_named['respiratory_rate'] = df_named['respiratory_rate'].fillna(df_without_outliers['respiratory_rate'].mean())
```

Hot extremities should correlate with an elevated rectal temp. normal temp is 37.8
Высокой ректальной темп будем считать от 38. Для высокой темп конечностей заполним ректальную температуру средним по повышенной темп, для повышенной темп заполним температуру конечностей 2 = Warm.
Нормальной температурой будем считать от 37 до 38, для нее температуру конечностей ставим 1 = Normal


```python
def temp_classification(temp):
    '''классифицирует температуру на высокую, нормальную, низкую и добавляет соотв. столбец'''
    t_class = np.NaN
    if temp >= 38:
        t_class = 'high'
    if temp < 37:
        t_class = 'low'
    if temp>= 37 and temp <38:
        t_class = 'normal'
    return t_class    
        
```

```python
df_named['temp_class'] = df_named['rectal_temperature'].apply(temp_classification)
df_named
```

```python
df_without_outliers = remove_outliers('rectal_temperature', df_named)
df_without_outliers.describe()
```

```python
df_without_outliers.groupby(['temp_class'])['temperature_of_extremities'].agg(pd.Series.mode)
```

Видим, что самые частые значения не соответствуют нашему предположению - при высокой температуре конечности чаще прохладные, а при низкой - теплые. Будем заполнять в итоге модой (которой почему-то нет в методе groupby)

```python
def fill_temp_ex(row):
        if row['temp_class'] == 'high':
            return 3.0
        if row['temp_class'] == 'low':
            return 2.0   
        if row['temp_class'] == 'normal':
            return 1.0  
        else:
            return np.NaN
```

```python
df_named['temperature_of_extremities'] = df_named.fillna(df_named.apply(fill_temp_ex, axis = 1))
```

Разделим пульс на категории: low - <30, normal - 30-40, medhigh - 40-90, high - >90

```python
def pulse_classification(pulse):
    '''делит пульс на категории: low - <30, normal - 30-40, medhigh - 40-90, high - >90'''
    p_class = np.NaN
    if pulse >=90:
        p_class = 'high'
    if pulse < 30:
        p_class = 'low'
    if pulse >= 30 and pulse <=40:
        p_class = 'normal'
    if pulse > 40 and pulse <90:
        p_class = 'medhigh'    
    return p_class    
```

```python
df_named['pulse_class'] = df_named['pulse'].apply(pulse_classification)
df_named
```

```python
df_without_outliers = remove_outliers('pulse', df_named)
```
